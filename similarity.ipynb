```python
import pandas as pd
similarity=pd.read_csv('copy.csv')
```

```python
similarity.shape
```

```python
q=similarity[similarity['name'].str.contains('query')]
k=similarity[similarity['name'].str.contains('key')]
v=similarity[similarity['name'].str.contains('value')]
print(q.shape,k.shape,v.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x=np.arange(12)
fig = plt.figure(figsize=(10,6))
plt.xticks(x)
plt.plot(x,q['sim'])
```

```python
x=np.arange(12)
fig = plt.figure(figsize=(10,6))
plt.xticks(x)
plt.plot(x,k['sim'])
```

```python
x=np.arange(12)
fig = plt.figure(figsize=(10,6))
plt.xticks(x)
plt.plot(x,v['sim'])
```

```python
x=np.arange(12)
fig = plt.figure(figsize=(10,6))
plt.xticks(x)
plt.plot(x,q['sim'],label='q')
plt.xticks(x)
plt.plot(x,k['sim'],label='k')
plt.xticks(x)
plt.plot(x,v['sim'],label='v')
plt.legend()
```
